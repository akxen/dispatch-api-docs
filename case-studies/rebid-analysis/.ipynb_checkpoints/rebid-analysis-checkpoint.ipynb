# Rebid Analysis
The following sections demonstrate how the NEMDE API can be used estimate the contribution of a rebid to a change in dispatch prices.

## Overview
On 27 November 2020 at 4.15pm the dispatch price in NSW was approximately 96 $/MWh. Dispatch prices increased to 290 $/MWh in the following interval at 4.20pm before falling to 93 $/MWh at 4.25pm. Approximately 2 minutes before the interval in which the price spike occurred a unit at Bayswater power station (trader ID BW01) submitted a rebid which indicated the unit's max available capacity was 130 MW less than previously anticipated. The following figure plots the power output and max availability for this unit, along with NSW dispatch prices.

````python
import os

import pandas as pd
import matplotlib.dates as md
import matplotlib.pyplot as plt
import IPython.display as display


def get_dispatch_scada(data_dir, filename):
    """Get dispatch SCADA data"""

    df = pd.read_csv(os.path.join(data_dir, filename), skiprows=1)
    df = df.iloc[:-1]
    df['SETTLEMENTDATE'] = pd.to_datetime(df['SETTLEMENTDATE'])

    return df


def get_dispatch_scada_window(data_dir, filename, duid, timestamp, window):
    """Get DUID dispatch over a given window"""

    df = get_dispatch_scada(data_dir=data_dir, filename=filename)

    start = pd.to_datetime(timestamp) - pd.Timedelta(minutes=5*window)
    end = pd.to_datetime(timestamp) + pd.Timedelta(minutes=5*window)

    mask = ((df['SETTLEMENTDATE'] >= start)
            & (df['SETTLEMENTDATE'] <= end) & (df['DUID'] == duid))

    scada = (df.loc[mask, :].set_index('SETTLEMENTDATE')['SCADAVALUE']
             .sort_index())

    return scada


def get_region_price(data_dir, filename):
    """Get region prices"""

    df = pd.read_csv(os.path.join(data_dir, filename), skiprows=1)
    df = df.iloc[:-1]

    df['SETTLEMENTDATE'] = pd.to_datetime(df['SETTLEMENTDATE'])

    prices = (df.sort_values(by=['SETTLEMENTDATE', 'REGIONID', 'INTERVENTION'])
              .drop_duplicates(subset=['SETTLEMENTDATE', 'REGIONID'], keep='last'))

    return prices.set_index(['SETTLEMENTDATE', 'REGIONID'])['RRP']


def get_region_price_window(data_dir, filename, duid, timestamp, window):
    """Get region price window"""

    df = get_region_price(data_dir=data_dir, filename=filename)

    start = pd.to_datetime(timestamp) - pd.Timedelta(minutes=5*window)
    end = pd.to_datetime(timestamp) + pd.Timedelta(minutes=5*window)

    prices = df.unstack()
    mask = (prices.index >= start) & (prices.index <= end)

    return prices.loc[mask, :]


def get_bidperoffer(data_dir, tmp_dir, filename, bid_type, use_cache=True):
    """Get bidperoffer (quantity bands, max avail, and ramp rates)"""

    output_filename = f'{bid_type}.csv'
    if use_cache:
        try:
            df = pd.read_csv(os.path.join(tmp_dir, output_filename))
            df['INTERVAL_DATETIME'] = pd.to_datetime(df['INTERVAL_DATETIME'])
            df['OFFERDATE'] = pd.to_datetime(df['OFFERDATE'])
            df['SETTLEMENTDATE'] = pd.to_datetime(df['SETTLEMENTDATE'])
            return df
        except FileNotFoundError:
            pass

    # Read data from file
    df = pd.read_csv(os.path.join(data_dir, filename), skiprows=1)
    df = df.iloc[:-1]

    # Filter by bid type (for testing)
    mask = df['BIDTYPE'] == bid_type
    df = df.loc[mask, :]
    df['INTERVAL_DATETIME'] = pd.to_datetime(df['INTERVAL_DATETIME'])
    df['OFFERDATE'] = pd.to_datetime(df['OFFERDATE'])
    df['SETTLEMENTDATE'] = pd.to_datetime(df['SETTLEMENTDATE'])

    df.to_csv(os.path.join(tmp_dir, output_filename), index=False)

    return df


def get_bidperoffer_window(data_dir, tmp_dir, filename, duid, bid_type, timestamp, window, use_cache=True):
    """Get DUID bid information over a given window"""

    df = get_bidperoffer(data_dir=data_dir, tmp_dir=tmp_dir,
                         filename=filename, bid_type=bid_type, use_cache=use_cache)

    start = pd.to_datetime(timestamp) - pd.Timedelta(minutes=5*window)
    end = pd.to_datetime(timestamp) + pd.Timedelta(minutes=5*window)

    mask = (df['INTERVAL_DATETIME'] >= start) & (
        df['INTERVAL_DATETIME'] <= end) & (df['DUID'] == duid)
    bids = df.loc[mask, :].set_index('INTERVAL_DATETIME').sort_index()

    return bids

data_directory = os.path.join(os.path.curdir, 'data')
tmp_directory = os.path.join(os.path.curdir, 'tmp')

biddayoffer_filename = 'PUBLIC_DVD_BIDDAYOFFER_202011010000.CSV'
bidperoffer_filename = 'PUBLIC_DVD_BIDPEROFFER_D_202011010000.CSV'
region_price_filename = 'PUBLIC_DVD_DISPATCHPRICE_202011010000.CSV'
dudetail_filename = 'PUBLIC_DVD_DUDETAILSUMMARY_202011010000.CSV'
scada_filename = 'PUBLIC_DVD_DISPATCH_UNIT_SCADA_202011010000.CSV'

timestamp = '2020-11-27 16:20:00'
duid = 'BW01'
window = 12
use_cache = True
bid_type = 'ENERGY'
trade_type = 'ENOF'

prices = get_region_price_window(
    data_dir=data_directory,
    filename=region_price_filename,
    duid=duid,
    timestamp=timestamp,
    window=window)

scada = get_dispatch_scada_window(
    data_dir=data_directory,
    filename=scada_filename,
    duid=duid,
    timestamp=timestamp,
    window=window
)

bids = get_bidperoffer_window(
    data_dir=data_directory,
    tmp_dir=tmp_directory,
    filename=bidperoffer_filename,
    duid=duid,
    bid_type=bid_type,
    timestamp=timestamp,
    window=window,
    use_cache=use_cache
)
````

````python
from matplotlib.ticker import MultipleLocator

def plot_window(scada, bids, prices):
    """Plot prices, dispatch, and max avail for BW01 over given window"""

    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    line1 = ax1.plot(scada.index, scada.values, linestyle='--', color='#d60e00', linewidth=1.5, alpha=0.8, label='BW01 output', marker='o', fillstyle='none')
    line2 = ax1.plot(bids.index, bids['MAXAVAIL'].values, color='#0026d1', linewidth=1.5, alpha=0.8, label='BW01 MaxAvail', marker='o', fillstyle='none')
    line3 = ax2.plot(prices.index, prices['NSW1'].values, color='#d47800', linewidth=1.5, alpha=0.8, label='NSW price', marker='o', fillstyle='none')
    ax1.set_xlabel('Time', fontsize=18, fontweight='bold')
    ax1.set_ylabel('Output / MaxAvail (MW)', fontsize=18, fontweight='bold')
    ax2.set_ylabel('Dispatch price ($/MWh)', fontsize=18, fontweight='bold')
    ax1.xaxis.set_minor_locator(md.MinuteLocator(byminute=range(0, 60, 5)))
    ax1.xaxis.set_major_locator(md.MinuteLocator(byminute=[0, 15, 30, 45]))
    ax1.xaxis.set_major_formatter(md.DateFormatter('%H:%M'))
    ax1.text(0.01, .98, pd.Timestamp(timestamp).strftime('%Y/%m/%d'), transform=ax1.transAxes, color='black', verticalalignment='top', fontsize=15, fontweight='bold')
    lines = line1 + line2 + line3
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc=2, fontsize=15, bbox_to_anchor=(0.75, 0.97), framealpha=0.6, frameon=False)
    ax1.tick_params(labelsize=15)
    ax2.tick_params(labelsize=15)
    ax1.yaxis.set_major_locator(MultipleLocator(50))
    ax1.yaxis.set_minor_locator(MultipleLocator(10))
    ax2.yaxis.set_major_locator(MultipleLocator(50))
    ax2.yaxis.set_minor_locator(MultipleLocator(10))
    fig.set_size_inches(14, 7)
    ax1.grid(linestyle='--', linewidth=0.7, alpha=0.8)

    plt.show()

plot_window(scada=scada, bids=bids, prices=prices)
````

When analysing historical data it is difficult to estimate the extent to which a change in max availbility for BW01 contributed to the price movement at 4.20pm. The NEMDE API addresses this limitation by allowing users to estimate dispatch outcomes under counterfactual scenarios. For instance, a scenario can be constructed in which BW01 does not reduce its max availability at 4.20pm. Dispatch prices for this scenario can then be compared to historical outputs. As all other system parameters are kept constant, it is possible to isolate the rebid's contribution to the change in dispatch prices.

## Aims and scope
The following analysis investigates how the rebid submitted by BW01 impacted dispatch prices observed at 4.20pm on 27 Novermber 2020. First, model validation is undertaken to assess the NEMDE API's ability to emulate NEMDE outputs for the interval under investigation. If close correspondence is observed between NEMDE API and NEMDE solutions then counterfactual scenarios will be constructed to estimate dispatch prices that would have arisen if BW01 had not rebid its capacity. 

## Validation
Historical NEMDE case files include the outputs returned by NEMDE when run for a given set of inputs describing the NEM's state. Model validation is performed by passing the same parameters to the NEMDE API as were passed to NEMDE, and comparing the NEMDE API's solution to historical NEMDE outputs. The degree to which solutions obtained from the NEMDE API correspond with NEMDE outputs indicates the NEMDE API's ability to emulate NEMDE for the dispatch interval under investigation. If close correspondence between the two solutions is observed then it is appropriate to proceed with the counterfactual scenario analysis.

The following sections summarise the difference between the results obtained from the NEMDE API and historical NEMDE outputs for the dispatch interval ending at 4.20pm on 27 November 2020 - the interval for which the price spike is observed. 

### Trader solution
Summary statistics are computed for the difference between the NEMDE API's solution and NEMDE outputs, with the difference defined as follows: 

```
Solution Difference = NEMDE API Solution - NEMDE Solution.
```

Statistics for the difference between the trader solution obtained from the NEMDE API and NEMDE are shown below.

````python
import time

import requests
from dotenv import load_dotenv
from jsonpath_ng import jsonpath
from jsonpath_ng.ext import parse


def timestamp_to_case_id(timestamp):
    """Convert timestamp to case ID"""

    # Get case ID
    ts = pd.to_datetime(timestamp)

    # Assumed start of trading day
    start = pd.to_datetime(f'{ts.year}-{ts.month}-{ts.day} 04:00:00')

    # Check if trading day is actually previous calendar day
    t_diff = ts - start
    if t_diff.total_seconds() < 0:
        start = start - pd.Timedelta(days=1)
        t_diff = ts - start

    # Intervals from start of trading day
    intervals = int(t_diff.total_seconds() / 300)

    # Case ID corresponding to timestamp
    return f'{start.year}{start.month:02}{start.day:02}{intervals:03}'


def download_casefile(base_url, headers, case_id):
    """
    Download a case file

    Parameters
    ----------
    base_url : str
        Base URL for the NEMDE API

    headers : dict
        Contains auth token used to authenticate requests

    case_id : str
        Case ID to download e.g. '20201101001'. The format is as follows
        {year}{month}{day}{interval_id} where interval_id is in [001, ..., 288]

    Returns
    -------
    casefile : dict
        Parameters describing the system's state
    """

    url = base_url + f'data/casefile/{case_id}'
    response = requests.get(url=url, headers=headers)
    casefile = response.json()

    return casefile


def submit_casefile(base_url, headers, casefile):
    """Submit case file to the job queue"""

    # Construct request body and URL
    body = {'casefile': casefile}
    url = base_url + 'jobs/create'

    # Send job to queue and return job meta data
    response = requests.post(url=url, headers=headers, json=body)

    return response.json()


def submit_casefile_validation(base_url, headers, casefile):
    """
    Submit case file to the job queue. Results will compare NEMDE solution
    to NEMDE API's solution
    """

    # Construct request body and URL
    body = {
        'casefile': casefile, 
        'options': {'solution_format': 'validation'},
    }
    url = base_url + 'jobs/create'

    # Send job to queue and return job meta data
    response = requests.post(url=url, headers=headers, json=body)

    return response.json()


def check_job_status(base_url, headers, job_id):
    """Check job status given a job ID"""

    url = base_url + f'jobs/{job_id}/status'

    return requests.get(url=url, headers=headers).json()


def get_job_results(base_url, headers, job_id):
    """Extract job results from the queue"""

    url = base_url + f'jobs/{job_id}/results'
    response = requests.get(url=url, headers=headers)

    return response.json()


def get_validation_results(base_url, headers, case_id):
    """Compare NEMDE and NEMDE API solution for given interval"""

    # Download case file
    casefile = download_casefile(base_url=base_url, headers=headers, case_id=case_id)

    # Check prices
    job_info = submit_casefile_validation(base_url=base_url, headers=headers, casefile=casefile)
    job_id = job_info.get('job_id')
    job_status = check_job_status(base_url=base_url, headers=headers, job_id=job_id)
    
    while job_status['status'] != 'finished':
        time.sleep(6)
        job_status = check_job_status(base_url=base_url, headers=headers, job_id=job_id)
#         print(job_status['status'])

    results = get_job_results(base_url=base_url, headers=headers, job_id=job_id)
    
    return results

load_dotenv('config/.env')

TOKEN = os.getenv('TOKEN')
headers = {'Authorization': f'Token {TOKEN}'}

# Base URL endpoint for the NEMDE API
# base_url = 'https://dispatch.envector.com/api/v1/'
base_url = 'http://dispatch.localhost/api/v1/'

results = get_validation_results(base_url=base_url, headers=headers, case_id=timestamp_to_case_id(timestamp=timestamp))
````

````python
def get_solution_summary(results, key):
    """
    Compute summary statistics for difference between NEMDE and NEMDE API 
    solution attributes
    """
    
    df = pd.DataFrame(results['results']['output'][key])
    df['difference'] = df['model'] - df['actual']
    summary = df.groupby('key').apply(lambda x: x['difference'].describe())
    
    # Convert to markdown
    markdown = summary.to_markdown()
    
    return display.Markdown(markdown)

get_solution_summary(results=results, key='TraderSolution')
````

The following figures visualise the correspondence between the dispatch targets obtained from the NEMDE API and those returned by NEMDE for each trade type. Each point represents a dispatch target for a given trader. The horizontal axis denotes the solution obtained from the NEMDE API, while the vertical axis denotes the solution obtained from NEMDE. The dashed line has a slope of one, and indicates perfect correspondence between the NEMDE API and NEMDE solutions.

````python
def plot_dispatch_targets(ax, results, key):
    """Plot dispatch targets on specified axis"""
    
    df = pd.DataFrame(results['results']['output']['TraderSolution'])
    df = df.loc[df['key'] == key, :]

    min_val = min(df['model'].min(), df['actual'].min())
    max_val = max(df['model'].max(), df['actual'].max())
    ax.scatter(x=df['model'], y=df['actual'].values, alpha=0.7, color='#d6001d')
    ax.plot([min_val, max_val], [min_val, max_val], linestyle='--', color='black', alpha=0.8, linewidth=0.8)
    ax.set_title(key, fontsize=16)
    ax.set_xlabel('Model (MW)', fontsize=14)
    ax.set_ylabel('NEMDE (MW)', fontsize=14)
    
    return ax


def plot_dispatch_target_comparison(results):
    """Plot dispatch targets for each trade type"""
    
    fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7, ax8), (ax9, ax10)) = plt.subplots(nrows=5, ncols=2)

    ax1 = plot_dispatch_targets(ax=ax1, results=results, key='@R6Target')
    ax2 = plot_dispatch_targets(ax=ax2, results=results, key='@L6Target')
    ax3 = plot_dispatch_targets(ax=ax3, results=results, key='@R60Target')
    ax4 = plot_dispatch_targets(ax=ax4, results=results, key='@L60Target')
    ax5 = plot_dispatch_targets(ax=ax5, results=results, key='@R5Target')
    ax6 = plot_dispatch_targets(ax=ax6, results=results, key='@L5Target')
    ax7 = plot_dispatch_targets(ax=ax7, results=results, key='@R5RegTarget')
    ax8 = plot_dispatch_targets(ax=ax8, results=results, key='@L5RegTarget')
    ax9 = plot_dispatch_targets(ax=ax9, results=results, key='@EnergyTarget')
    ax10.axis('off')

    fig.set_size_inches(10, 20)
    fig.subplots_adjust(hspace=0.4)


plot_dispatch_target_comparison(results=results)
plt.show()
````

There is excellent correpondence between the two models with respect to dispatch targets. The symmetric dispersion of points around the dashed line for the L6Target is likely due to the absence of tie-breaking constraints for FCAS offers.

## Contribution of rebid to region price change
The bid for BW01 at 4.15pm is compared with the bid at 4.20pm in the following table.

````python
def get_bid_parameters(bids, timestamp):
    """Get bid parameters for a given """
    
    # Convert to timestamp object
    ts = pd.Timestamp(timestamp)
    
    # Timestamp for preceding interval
    ts_minus_one = ts - pd.Timedelta(minutes=5)
    
    columns = ['MAXAVAIL'] + [f'BANDAVAIL{i}' for i in range(1, 11)]
    
    # Only retain selected timestamps and compute difference
    df = bids.loc[[ts_minus_one, ts], columns].T
    df['Difference (MW)'] = df.loc[:, ts] - df.loc[:, ts_minus_one]
    df.columns.name = None
    
    markdown = df.to_markdown()
    
    return display.Markdown(markdown)

get_bid_parameters(bids=bids, timestamp=timestamp)
````

The case file corresponding to the 4.20pm dispatch interval is updated such that the bid for BW01 applying at 4.15pm also applies at 4.20pm (the MaxAvail parameter for BW01 is set to 590MW for the 4.20pm case file). The NEMDE API is used to run the augmented case file, with region solution metrics corresponding to this scenario shown in the following table.

````python
def get_casefile_parameter(casefile, expression):
    """
    Get parameter given a case file and JSON path expression

    Parameters
    ----------
    casefile : dict
        System parameters

    expression : str
        JSON path expression to value or object that should be
        extracted

    Returns
    -------
    Value corresponding to JSON path expression.
    """

    jsonpath_expr = parse(expression)
    values = [match.value for match in jsonpath_expr.find(casefile)]

    # Check only one match found
    if len(values) != 1:
        raise Exception(f'Expected 1 match, encountered {len(values)}')

    return values[0]


def update_casefile_parameter(casefile, expression, new_value):
    """
    Update case file parameter

    Parameters
    ----------
    casefile : dict
        System parameters

    expression : str
        JSON path to value or object that should be updated

    new_value : str, float, or int
        New value for parameter

    Returns
    -------
    casefile : dict
        Updated case file
    """

    jsonpath_expr = parse(expression)
    values = [match.value for match in jsonpath_expr.find(casefile)]

    # Check only one match found
    if len(values) != 1:
        raise Exception(f'Expected 1 match, encountered {len(values)}')

    # Update case file
    jsonpath_expr.update(casefile, new_value)

    return casefile


def get_bid_parameters(base_url, headers, timestamp, trader_id, trade_type):
    """Get bid parameters for a given trader and trade type"""
    
    case_id = timestamp_to_case_id(timestamp=timestamp)
    casefile = download_casefile(base_url=base_url, headers=headers, case_id=case_id)
    
    expressions = ([f"""NEMSPDCaseFile.NemSpdInputs.PeriodCollection.Period.TraderPeriodCollection.TraderPeriod[?(@TraderID='{trader_id}')].TradeCollection.Trade[?(@TradeType='{trade_type}')].@MaxAvail"""]
               + [f"""NEMSPDCaseFile.NemSpdInputs.PeriodCollection.Period.TraderPeriodCollection.TraderPeriod[?(@TraderID='{trader_id}')].TradeCollection.Trade[?(@TradeType='{trade_type}')].@BandAvail{i}""" for i in range(1, 11)])

    parameters = {
        i: get_casefile_parameter(casefile=casefile, expression=i) for i in expressions
    }
    
    return parameters


def run_scenario(base_url, headers, timestamp, updates):
    """Run scenario with specified updates"""
    
    case_id = timestamp_to_case_id(timestamp=timestamp)
    casefile = download_casefile(base_url=base_url, headers=headers, case_id=case_id)
    
    # Apply updates
    for i, j in updates.items():
        casefile = update_casefile_parameter(casefile=casefile, expression=i, new_value=j)
    
    # Check prices
    job_info = submit_casefile(base_url=base_url, headers=headers, casefile=casefile)
    job_id = job_info.get('job_id')
    job_status = check_job_status(base_url=base_url, headers=headers, job_id=job_id)
    
    while job_status['status'] != 'finished':
        time.sleep(6)
        job_status = check_job_status(base_url=base_url, headers=headers, job_id=job_id)
#         print(job_status['status'])

    results = get_job_results(base_url=base_url, headers=headers, job_id=job_id)
    
    return results


# Updates to apply to formulate couterfactual scenario
bid_parameter_updates = get_bid_parameters(
    base_url=base_url,
    headers=headers,
    timestamp=pd.Timestamp(timestamp) - pd.Timedelta(minutes=5),
    trader_id=duid,
    trade_type=trade_type
)

# Base case - no updates
base = run_scenario(
    base_url=base_url,
    headers=headers,
    timestamp=timestamp,
    updates=dict()
)

# Counterfactual scenario - apply BW01 bids from previous interval
scenario = run_scenario(
    base_url=base_url,
    headers=headers,
    timestamp=timestamp,
    updates=bid_parameter_updates
)
````

````python
scenario_region = pd.DataFrame(scenario['results']['output']['RegionSolution']).set_index('@RegionID').drop(['@CaseID', '@Intervention'], axis=1)
display.Markdown(scenario_region.to_markdown())
````

The NEMDE API is then used to run the base case scenario in which no changes are made to the historical case file. Note excellent correspondence was observed between NEMDE and NEMDE API energy prices in the model validation section. Region solution results for the base case are shown below.

````python
base_region = pd.DataFrame(base['results']['output']['RegionSolution']).set_index('@RegionID').drop(['@CaseID', '@Intervention'], axis=1)
display.Markdown(base_region.to_markdown())
````

The difference between the counterfactual scenario and the base case is as follows.

````python
display.Markdown(scenario_region.subtract(base_region).to_markdown())
````

The analysis suggests the rebid led to slightly higher prices in New South Wales and Queensland, but reduced prices in Tasmania. These price differences are small relative to the magnitude of the price spike, suggesting the rebid made by BW01 was not a determinative factor in the large price change observed.

## Summary
The NEMDE API has been used to estimate dispatch prices that would arise if BW01 had not rebid its max availability. Model validation is first performed to assess the NEMDE API's ability to emulate NEMDE outputs for the dispatch interval in question, with close correspondence between the NEMDE API's solution and NEMDE outputs observed. A counterfactual scenario is then constructed which assumes no reduction in max availablity for BW01 at 4.20pm. This scenario is compared with the base case in which BW01 does rebid its capacity, with the results suggesting the rebid has a negligible impact on prices at 4:20pm - it is likely other factors, unrelated to the rebid submitted by BW01, are driving the price movement in this dispatch interval.
